```python
wide_df= pd.DataFrame({
    "Name":["Ram","shyam"],
    "science":[34,56],
    "math":[34,67],
    "english":[64,74]

})
print(wide_df)
```

```python
long_df=pd.melt(
    frame=wide_df,
    id_vars=["Name"],
    var_name="subject",
    value_name="marks"
)
print(long_df)
```

```python
new_df = long_df.pivot(
    index="Name",
    columns="subject",values="marks"
)
new_df
```

```python
stacked_df= long_df.stack()
print(stacked_df)
```

```python
unstacked_df=stacked_df.unstack()
print(unstacked_df)
```

```python
pivot_table_df=pd.pivot_table(
    data=long_df,
    index="Name",

    columns="subject",
    values="marks",
    aggfunc="median"
)
print(pivot_table_df)
```
