```python
import pandas as pd
df= pd.DataFrame({
    "Date":['2023-3-4','2023-3-25','2023-3-6','2023-3-9'],
    "sales":[132,354,56,676]
})
print(df.dtypes)
```

```python
df['Date']=pd.to_datetime(df['Date'])
print(df.dtypes)
```

```python
df = df.set_index("Date")
print(df.head())
```

```python
df["Year"]=df.index.year
df["Month"]=df.index.month
df["Day"]=df.index.day
df["Day_name"]=df.index.day_name()
df['Donth_name']=df.index.month_name()
```
