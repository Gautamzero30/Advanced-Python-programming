```python
import pandas as pd
df= pd.DataFrame ({
    "eid":[101,232,4545],
    "name":['hello',"stop","taxi"],
    "level":["assistant","professor","lecturer"]
})
df.dtypes
```

```python
df["level"]=df["level"].astype("category")
df.dtypes
```

```python
df['level']=df['level'].cat.rename_categories({
    "assistant":"ast",
    "professor":"proff"
})
df.head()
```

```python
df['level_code']=df['level'].cat.codes
print(df.head())
```

```python
encoded= pd.get_dummies(df["level"],prefix="lev")
final_df=pd.concat(objs=[df,encoded],axis=1)
print(final_df)
```
